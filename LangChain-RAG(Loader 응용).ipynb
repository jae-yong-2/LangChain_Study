```python
# WebBaseLoader

import bs4
from langchain_community.document_loaders import WebBaseLoader

url_one = "https://blog.langchain.dev/customers-replit/"
url_two = "https://blog.langchain.dev/langgraph-v0-2/"

loader = WebBaseLoader(
    web_paths=(url_one, url_two),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("article-header", "article-content")
            )
      )
    )

docs = loader.load()
docs[0]
```
